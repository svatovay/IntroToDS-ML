```python
import pandas as pd
import numpy as np
```

```python
students_performance = pd.read_csv('StudentsPerformance.csv')
```

```python
students_performance.head()
```

```python
students_performance.loc[students_performance.gender == 'female', ['gender', 'writing score']]
```

```python
mean_writing_score = students_performance['writing score'].mean()
```

```python
students_performance.loc[students_performance['writing score'] > mean_writing_score]
```

```python
students_performance.loc[(students_performance['writing score'] > mean_writing_score) & (students_performance.gender == 'female')]
```

```python
query = (students_performance['lunch'] == 'free/reduced')
len(students_performance.loc[query].index) / len(students_performance.index)
```

```python
students_performance.describe()
```

```python
students_performance.groupby('lunch').aggregate({'math score':['count', 'mean', 'std'], 'reading score':['count', 'mean', 'std'], 'writing score' :['count', 'mean', 'std']})
```

```python
new_name_map = {c_name: c_name.replace(' ', '_') for c_name in students_performance.columns}
students_performance = students_performance.rename(columns = new_name_map)
students_performance.head()
```

```python
students_performance.query('writing_score > 74')
```

```python
students_performance.query('gender == "female" & writing_score > 78')
```

```python
writing_score_query = 90
students_performance.query('writing_score > @writing_score_query')
```

```python
score_columns = [_ for _ in list(students_performance) if _.endswith('score')]
students_performance[score_columns]
```

```python
students_performance.filter(like='score')
```
