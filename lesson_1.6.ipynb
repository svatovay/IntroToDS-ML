```python
import pandas as pd
import numpy as np
```

```python
students_performance = pd.read_csv('StudentsPerformance.csv')
```

```python
students_performance.head()
```

```python
mean_scores = students_performance.groupby(['gender', 'race/ethnicity'], as_index=False) \
.aggregate({'math score': 'mean', 'reading score': 'mean',}) \
.rename(columns = {'math score': 'mean_math_score', 'reading score': 'mean_reading_score'})
```

```python
# mean_scores.loc[[('female', 'group A'), ('female', 'group B')]]
```

```python
students_performance.sort_values(['gender', 'math score'], ascending=False) \
.groupby('gender').head(5)
```

```python
students_performance['total_score'] = students_performance['math score'] + students_performance['reading score'] + students_performance['writing score']
```

```python
students_performance.head()
```

```python
students_performance = students_performance.assign(total_score_log = np.log(students_performance.total_score))
```

```python
concentrations = pd.read_csv('algae.csv')
mean_cols = {col: 'mean' for col in concentrations.columns[3:]}
mean_concentrations = concentrations.groupby('genus').aggregate(mean_cols)
mean_concentrations.head()
```

```python
alanin_concentrations = concentrations.groupby('genus').aggregate({'alanin': ['min', 'mean', 'max']}).round(2)
alanin_concentrations.head()
```

```python
concentrations.head()
```

```python
groups_species_count = concentrations.groupby('group').aggregate({'species': 'count'})
groups_species_count
```

```python
groups_sucrose_range = concentrations.groupby('group').aggregate({'sucrose': ['min', 'max']})
groups_sucrose_range.assign(range=groups_sucrose_range['sucrose']['max'] - groups_sucrose_range['sucrose']['min']).round(2)
```
